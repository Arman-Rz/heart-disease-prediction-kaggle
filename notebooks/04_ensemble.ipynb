```python
from pathlib import Path
import pandas as pd
import numpy as np

from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import roc_auc_score
from sklearn.model_selection import train_test_split
from catboost import CatBoostClassifier, Pool
import lightgbm as lgb
from xgboost import XGBClassifier

import warnings
warnings.filterwarnings('ignore')

SEED = 42
```

```python
ROOT = Path.cwd().parent.resolve()
LGB = ROOT / "outputs/lightgbm/light_test.npy"
CAT = ROOT / "outputs/catboost/cat_test.npy"
XGB = ROOT / "outputs/xgboost/xgb_test.npy"
test_df = pd.read_csv(str(ROOT / "data/test.csv"), sep=',')

lgb_test = np.load(LGB)
cat_test = np.load(CAT)
xgb_test = np.load(XGB)

final_preds = ((0.1 * lgb_test) + (0.5 * xgb_test) + (0.4 * lgb_test)) / 3

submission = pd.DataFrame({
    "id": test_df["id"],
    "prediction": final_preds
})

submission.to_csv(ROOT / "outputs/submission.csv", index=False)
```
